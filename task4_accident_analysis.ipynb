```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("dataset/RTA Dataset.csv")
df.head()
```

```python
df.columns
```

```python
df['Hour'] = pd.to_datetime(
    df['Time'],
    format='%H:%M:%S',
    errors='coerce'
).dt.hour
```

```python
df[['Time', 'Hour']].head()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x='Hour', data=df)
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(
    y='Weather_conditions',
    data=df,
    order=df['Weather_conditions'].value_counts().index
)
plt.title("Accidents by Weather Conditions")
plt.xlabel("Number of Accidents")
plt.ylabel("Weather Condition")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(
    y='Road_surface_conditions',
    data=df,
    order=df['Road_surface_conditions'].value_counts().index
)
plt.title("Accidents by Road Surface Condition")
plt.xlabel("Number of Accidents")
plt.ylabel("Road Surface")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Accident_severity', data=df)
plt.title("Accident Severity Distribution")
plt.xlabel("Severity")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(
    y='Light_conditions',
    data=df,
    order=df['Light_conditions'].value_counts().index
)
plt.title("Accidents by Light Conditions")
plt.xlabel("Number of Accidents")
plt.ylabel("Light Condition")
plt.show()
```

### Key Observations
- Accidents are more frequent during peak traffic hours.
- Adverse weather conditions increase accident occurrence.
- Poor road surface conditions are associated with higher accident rates.
- Serious accidents form a significant portion of total cases.
- Visibility conditions affect accident frequency.
